```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import math
from scipy.integrate import trapz
from glob import glob
import os 
```

#####  Define Matplotlib Plotting Settings

```python
font = {'size'   : 15}

matplotlib.pyplot.rc('font', **font)
```

#### Import CSV files of fabric swatches from heated hot plate machine

```python
D200_swatches = glob('D200*')
D_thermal_swatches =glob('DT*')
PVC_swatches = glob('PVC*')
```

#### Initalize array for max flux from each replicate

```python
D200_max_flux = np.zeros(len(D200_swatches))
D_thermal_max_flux = np.zeros(len(D_thermal_swatches))
PVC_max_flux = np.zeros(len(PVC_swatches))

fabric_mean=[]
fabric_std = []
```

### Calculate Max Heat Flux from all Replicates

```python
fig, ax = plt.subplots(figsize=(12,7))
plt.xlabel('Time (seconds)', labelpad=20)
plt.ylabel(r'Heat Flux $\frac{W}{m^2}$')

for i, rep in enumerate(D200_swatches):
    df = pd.read_csv(rep, skiprows = 16)  # data starts at line 13
    reduced_df = df.loc[1:,['Time' ,'Heat Flux Measured']] # further reduce dataframe to include only the columns of interest 
    reduced_df = reduced_df.astype('float64')  # cast data from object ot float64
    D200_max_flux[i] = reduced_df['Heat Flux Measured'].max()  # return max value from series and append to my list
    
    label = f'Replicate {i+1}: ' # generate labels
    plt.plot(reduced_df['Time'], reduced_df['Heat Flux Measured'], label=f'{label + rep}') # plot each replicate stacked ontop
    
plt.legend()
plt.xlim((150, 450))  # adjust limits of x-axis
plt.title('D200 Fabric Heat Flux Experiment')

fabric_mean.append(np.mean(D200_max_flux))
fabric_std.append(np.std(D200_max_flux))
```

```python

steady_state_threshold_array = np.arange(5,10)
true_heat_flux_array = []

for steady_state_theshold in steady_state_threshold_array:

    df = pd.read_csv('PVC-1.csv', skiprows = 16)  # data starts at line 13
    reduced_df = df.loc[1:,['Time' ,'Heat Flux Measured']] # further reduce dataframe to include only the columns of interest 
    reduced_df = reduced_df.astype('float64')  # cast data from object ot float64


    dh = np.diff(reduced_df['Heat Flux Measured']) / np.diff(reduced_df['Time'])  # numerical Derivative of flux and time
    dh_and_time = np.zeros([len(dh), 2])  # intialie aray

    dh_and_time[:,0] = reduced_df.iloc[1:]['Time'].values  # populate array with time
    dh_and_time[:, 1] = dh  # populate array with derivative



    dh_and_time = dh_and_time[dh_and_time[:,1] > 0]  # remove anywhere derivative is 0 to remove step noise




    plt.plot(dh_and_time[:,0], dh_and_time[:,1])
    plt.ylabel('Change of Heat Flux ' '\n' r'with Respect to Time $\frac{W}{m^2s}$')
    plt.xlim(150, 525)

    threshold_delta = 100  # minimum step change in flux where event occurs 
    thershold_time = 0  # time where step change occured
    threshold_index = 0  # index in array where step change occured

    for i, derivative in enumerate(dh_and_time[:,1]):  # search where the first occurance of large step change in heat flux
        if derivative > threshold_delta:
            threshold_time = dh_and_time[i - steady_state_theshold,0]
            threshold_index = i - steady_state_theshold
            break

    # search where the first occurance of large step change in heat flux in reverse to denote return to Steady State

    for i, derivative in enumerate(np.flip(dh_and_time[:,1], 0)): # search the reverse direction for step change
        if derivative > threshold_delta:
            threshold_time_stop = dh_and_time[-i + 1 + steady_state_theshold,0]  # return time using negative index
            threshold_index_stop = dh_and_time.shape[0] - i + 1 + steady_state_theshold
            break

    # print('End of to Steady-State:' , dh_and_time[threshold_index, 0])

    background_noise = dh_and_time[threshold_index-steady_state_theshold: threshold_index, 1].mean()  # steady state Noise

    time_interval = threshold_time_stop -  threshold_time

    step_change_df = reduced_df[(reduced_df['Time'] >= threshold_time) & (reduced_df['Time'] <= threshold_time_stop)]

    total_heat_flux = trapz(step_change_df['Heat Flux Measured'],step_change_df['Time'])
    average_baseline = (step_change_df['Heat Flux Measured'].values[0] + step_change_df['Heat Flux Measured'].values[-1]) * 0.5
    baseline_total_flux = average_baseline*time_interval

    true_flux = total_heat_flux - baseline_total_flux
    true_heat_flux_array.append(true_flux)
    
# print(step_change_df['Heat Flux Measured'].values[0],step_change_df['Heat Flux Measured'].values[-1] , average_baseline )
# print(threshold_time_stop, threshold_time)
# print(total_heat_flux - baseline_total_flux)  # Joules / m^2
# print(time_interval)

true_heat_flux_array =np.asarray(true_heat_flux_array)

print(np.std(true_heat_flux_array)/np.mean(true_heat_flux_array) * 100)

```
